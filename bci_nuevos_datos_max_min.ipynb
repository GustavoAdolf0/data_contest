```python
train = full[full['USO'] == "train"].copy(deep= True)
train.drop(["USO"], axis=1, inplace=True)

test  = full[full['USO'] == "test"].copy(deep= True)
test.drop(["USO"], axis=1, inplace=True)

# Split data into X and y
X = train.drop("target_mes", axis=1)

y = train["target_mes"]


# Split data into train and test sets
np.random.seed(42)

# Split into train & test set
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.15)
```

```python
!pip install catboost
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor(loss_function="RMSE", n_estimators=500)

model.fit(X_train, y_train, verbose=0)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
y_pred= model.predict(X_train)
y_pred_t= model.predict(X_test)

r2_model_train = r2_score(y_train, y_pred)
r2_model_test  = r2_score(y_test, y_pred_t)
print(r2_model_train, r2_model_test)

#0.6326276241379618 0.5661098272062307
```

```python
mae_train = mean_absolute_error(y_train, y_pred)
mae_test  = mean_absolute_error(y_test, y_pred_t)
print(mae_train, mae_test )
#25.427348752237396 25.872691080985366
#89.26221376293945 100.5187909516407
```
