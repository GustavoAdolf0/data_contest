```python
stores=pd.merge(stores, datos_provincia[["city","Poblaci√≥n(2010)","capital",	"den_provincia",	"cantones",	"pob_provincia2020",	"%_pob"]], on="city", how="left")
```

todav√≠a se puede mejorar, las ciudades y estados se deben pasar a one hot encoder, la poblaci√≥n y desidad se puede clasificar dejando marcas de clases o similares

```python

du_stores= stores.copy(deep=False)
du_stores.head(2)
```

```python
stores["city-state"] = stores["city"] + "-" + stores["state"]
stores['cluster'] = stores['cluster'].astype(str)
stores["cluster"].replace({"5":  "peq",
                           "12": "peq",
                           "16": "peq",
                           "17": "peq"}, inplace=True)

stores["cluster"] = stores["cluster"] + "-cluster"
stores["type"] = stores["type"] + "-type"

stores["state"].replace({"Esmeraldas": "pob_peq",
                         "Chimborazo": "pob_peq", 
                         "Loja":"pob_peq",
                         "Imbabura":"pob_peq",
                         "Santa Elena":"pob_peq",
                         "Bolivar":"pob_peq",
                         "Pastaza":"pob_peq"}, inplace=True)

stores["city"].replace({"Esmeraldas": "city_peq",
                        "Riobamba": "city_peq", 
                        "Loja":"city_peq",
                        "Ibarra":"city_peq",
                        "Salinas":"city_peq",
                        "Bolivar":"city_peq",
                        "Pastaza":"city_peq"}, inplace=True)

stores["city-state"] = stores["city"] + "-" + stores["state"]
stores.head(2)
```

```python
stores.to_csv('drive/MyDrive/store_sales/stores_rev.csv')
```

```python
one_hot= pd.get_dummies(stores["city-state"])
stores=stores.join(one_hot)
```

```python
one_hot= pd.get_dummies(stores["state"])
stores=stores.join(one_hot)
```

```python
one_hot= pd.get_dummies(stores["cluster"])
stores=stores.join(one_hot)
```

```python
one_hot= pd.get_dummies(stores["type"])
stores=stores.join(one_hot)
```

```python
stores.head(2)
```

```python
#create the one hot encoder for state y city state
stores.drop(["city","state","city-state","cluster","type"],axis=1, inplace=True)
```
