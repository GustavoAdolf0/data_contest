```python
# Put models in a dictionary
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "Random Forest": RandomForestClassifier()}

# Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of differetn Scikit-Learn machine learning models
    X_train : training data (no labels)
    X_test : testing data (no labels)
    y_train : training labels
    y_test : test labels
    """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(X_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

n_estimators : This is the number of trees you want to build before taking the maximum voting or averages of predictions. Higher number of trees give you better performance but makes your code slower.

max_depth es la profundidad del último nodo de desición creado, el nodo raiz es la profundidad uno, en el tree 1, es de 3 dado que el pultimo en generar leaf es la en la tercera corrida hacia abajo.

min_sample_leaf es la cantidad mínima de observaciones que pueden entrar en un nodo

```python
# Create a hyperparameter grid for RandomForestClassifier
rf_grid = {"n_estimators": np.arange(10, 2500, 10),
           "max_depth": [ None, 10, 15, 20],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(2, 20, 2)}
```

CV es el hyperparamter para cros validation el modelo selecciona 1 de K para ser usado para entrenamiento y el k-esimo para usarlo en validación, de esta forma se evita el overfitting

```python
# Setup random seed
np.random.seed(42)

# Setup random hyperparameter search for RandomForestClassifier
rs_rf = RandomizedSearchCV(RandomForestClassifier(), 
                           param_distributions=rf_grid,
                           cv=5,
                           n_iter=500,
                           verbose=True)

# Fit random hyperparameter search model for RandomForestClassifier()
rs_rf.fit(X_train, y_train)
```

```python
def pre_pross(df):

  #this function grab from the file read to ready to ML
  df.loc[df["Sex"] =="male", "Sex_aux"] = 6
  df.loc[df["Sex"] == "female", "Sex_aux"] = 7 #create aux columns to 
  df["aux1"] = df["Pclass"]*df["Sex_aux"] #col aux to use in merge

  df= pd.merge(df, t_class, on="aux1", how= "left") #merge with t_class
  df= pd.merge(df, clases_aux, on="clasee", how= "left") #merge with clases to get age avg
  df.Age.fillna(train.Age_avg, inplace= True)
  df.Embarked.fillna("S", inplace=True)
  df.drop(['Cabin'], axis=1, inplace=True)
  df.drop(["Sex_aux","aux1" , "Age_avg" ], axis=1, inplace=True)

  one_hot= pd.get_dummies(train["Pclass"], prefix="Pclass")
  df=df.join(one_hot)

  one_hot= pd.get_dummies(train["Embarked"], prefix="Embarked")
  df=df.join(one_hot)

  one_hot= pd.get_dummies(train["clasee"])
  df=df.join(one_hot)

  one_hot= pd.get_dummies(train["Sex"])
  df=df.join(one_hot)

  df.drop(['Embarked',"Pclass","clasee", "Sex"], axis=1, inplace=True)

  miss = (df[df["Name"].str.contains("Miss")]).copy(deep=True)
  miss["miss"] = 1
  miss = miss[["PassengerId","miss"]]

  mister = (df[df["Name"].str.contains("Mr.")]).copy(deep=True)
  mister["mister"] = 1
  mister = mister[["PassengerId","mister"]]

  misteres = (df[df["Name"].str.contains("Mrs.")]).copy(deep=True)
  misteres["misteres"] = 1
  misteres = misteres[["PassengerId","misteres"]]

  rev = (df[df["Name"].str.contains("Rev.")]).copy(deep=True)
  rev["rev"] = 1
  rev = rev[["PassengerId","rev"]]

  df = pd.merge(df, miss, on="PassengerId", how= "left")
  df = pd.merge(df, mister, on="PassengerId", how= "left")
  df = pd.merge(df, misteres, on="PassengerId", how= "left")
  df = pd.merge(df, rev, on="PassengerId", how= "left")
  df.fillna(0, inplace= True)

  df.drop(['Name',"Ticket"], axis=1, inplace=True)
  df["fam"]= df["SibSp"] + df["Parch"]
  
  df = pd.merge(df, edad_class, on="Age", how= "left")

  df.set_index("PassengerId", inplace=True)
  return df
   

```
